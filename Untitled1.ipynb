```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('temperatures.csv')
```

```python
data
```

```python

data['Avg'] = (data['JAN-FEB']+data['MAR-MAY']+data['JUN-SEP']+data['OCT-DEC'])/12
```

```python
data
```

```python
X_train=data[:95]
X_test=data[95:]
```

```python
X = X_train['Avg'].values
Y = X_train['ANNUAL'].values
mean_x = np.mean(X)
mean_y = np.mean(Y)
```

```python
print("Printing mean x")
print(mean_x)
print("Printing mean y")
print(mean_y)
```

```python
m = len(X)
print("Number of samples in training set")
print(m)
numer = 0
denom = 0
for i in range(m):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2

b1 = numer / denom
b0 = mean_y - (b1 * mean_x)
print("Coefficient and bias is as follows")
print(b1, b0)
```

```python
max_x = np.max(X)
min_x = np.min(X)
```

```python
x = np.linspace(min_x, max_x)
y = b0 + b1 * x
```

```python
plt.plot(x, y, color='black', label='Regression Line')
plt.scatter(X, Y, c='yellow', label='Scatter Plot AVG VS ANNUAL')

plt.xlabel('AVG ')
plt.ylabel('ANNUAL')
plt.legend()
plt.show()
```

```python
sse = 0
for i in range(m):
    y_pred = b0 + b1 * X[i]   
    sse += (Y[i] - y_pred) ** 2    

print("Mean sqaure error of temperature  of train data",sse)
mse = sse/m
rmse = np.sqrt(mse)
print("Root Mean sqaure error is",rmse)
```

```python
ss_t = 0
ss_r = 0
for i in range(m):
    y_pred = b0 + b1 * X[i]
    ss_t += (Y[i] - mean_y) ** 2
    ss_r += (Y[i] - y_pred) ** 2
scorer2 = 1 - (ss_r/ss_t)
print("R^2 score for training data is",scorer2)
```
